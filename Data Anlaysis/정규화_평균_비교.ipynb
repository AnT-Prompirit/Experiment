```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load your datasets
# Assuming the paths are correct and pandas is correctly imported as pd
file_paths = [
    '/content/drive/MyDrive/앤트 공유 문서함/1. final result/ieapick/T2_stopwords_new.csv',
    '/content/drive/MyDrive/앤트 공유 문서함/1. final result/ieapick/T3_POS_new.csv',
    '/content/drive/MyDrive/앤트 공유 문서함/1. final result/ieapick/T5_T2_new.csv',
    '/content/drive/MyDrive/앤트 공유 문서함/1. final result/ieapick/T5_T3_new.csv',
]
datasets = [pd.read_csv(path) for path in file_paths]

# Step 2: Calculate mean values
mean_values = []
for df in datasets:
    mean_values.append({
        'ITA_mean': df['sd_ITA'].mean(),
        'IEA_mean': df['sd_IEA'].mean(),
        'Asthetic_mean': df['Asthetic'].mean(),
        'BERTScore_mean': df['BERTScore'].mean(),
    })

# Convert to DataFrame
mean_df = pd.DataFrame(mean_values, index=[ 'T2', 'T3', 'T5_T2', 'T5_T3'])

# Normalize each column to 0-1
normalized_df = (mean_df - mean_df.min()) / (mean_df.max() - mean_df.min())

# Calculate mean of each row
row_means = normalized_df.mean(axis=1)

mean_df['Norm_mean'] = row_means
mean_df

```
